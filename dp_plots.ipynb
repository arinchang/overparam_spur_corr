```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Weight Decay 0.0001 Varying Width Results with DP noise_mult = 1.0 and max_grad_clip = 1.0

```python
dp_1 = pd.read_csv('logs/test_dp_width1.csv')
dp_4 = pd.read_csv('logs/test_dp_width4.csv')
dp_16 = pd.read_csv('logs/test_dp_width16.csv')
dp_32 = pd.read_csv('logs/test_dp_width32.csv')
dp_64 = pd.read_csv('logs/test_dp_width64.csv')
dp_96 = pd.read_csv('logs/test_dp_width96.csv')
```

```python
w1_worst_group = np.mean(np.array(dp_1['avg_loss_group:3'])[40:])
w4_worst_group = np.mean(np.array(dp_4['avg_loss_group:3'])[40:])
w16_worst_group = np.mean(np.array(dp_16['avg_loss_group:3'])[40:])
w32_worst_group = np.mean(np.array(dp_32['avg_loss_group:3'])[40:])
w64_worst_group = np.mean(np.array(dp_64['avg_loss_group:3'])[40:])
w96_worst_group = np.mean(np.array(dp_96['avg_loss_group:3'])[40:])
```

```python
dp_1['avg_actual_loss'][49]
```

```python
test_errors = np.array([dp_1['avg_actual_loss'][49], dp_4['avg_actual_loss'][49], dp_16['avg_actual_loss'][49], dp_32['avg_actual_loss'][49], dp_64['avg_actual_loss'][49], dp_96['avg_actual_loss'][49]])
worst_group_errors = np.array([w1_worst_group, w4_worst_group, w16_worst_group, w32_worst_group, w64_worst_group, w96_worst_group])
plt.plot(np.array([1, 4, 16, 32, 64, 96]), test_errors, marker='o', label='average test')
plt.plot(np.array([1, 4, 16, 32, 64, 96]), worst_group_errors, marker='o', label='worst-group test')
plt.title("CelebA with 0.0001 Weight Decay and DP")
plt.xlabel("ResNet Width")
plt.ylabel("Error")
plt.legend()
```

```python
plt.plot(dp_32['avg_loss_group:0'], label='group 0')
plt.plot(dp_32['avg_loss_group:1'], label='group 1')
plt.plot(dp_32['avg_loss_group:2'], label='group 2')
plt.plot(dp_32['avg_loss_group:3'], label='group 3')
plt.title("celebA test error per group, width 32")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend()
```

```python
plt.plot(dp_64['avg_loss_group:0'], label='group 0')
plt.plot(dp_64['avg_loss_group:1'], label='group 1')
plt.plot(dp_64['avg_loss_group:2'], label='group 2')
plt.plot(dp_64['avg_loss_group:3'], label='group 3')
plt.title("celebA test error per group, width 64")
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend()
```

```python
dp_1_train = pd.read_csv("logs/train_dp_width1.csv")
dp_1_train = dp_1_train.groupby(['epoch']).mean()

dp_4_train = pd.read_csv("logs/train_dp_width4.csv")
dp_4_train = dp_4_train.groupby(['epoch']).mean()

dp_16_train = pd.read_csv("logs/train_dp_width16.csv")
dp_16_train = dp_16_train.groupby(['epoch']).mean()

dp_32_train = pd.read_csv("logs/train_dp_width32.csv")
dp_32_train = dp_32_train.groupby(['epoch']).mean()

dp_64_train = pd.read_csv("logs/train_dp_width64.csv")
dp_64_train = dp_64_train.groupby(['epoch']).mean()

dp_96_train = pd.read_csv("logs/train_dp_width96.csv")
dp_96_train = dp_96_train.groupby(['epoch']).mean()
```

```python
plt.plot(dp_1_train['avg_actual_loss'], label="width 1")
plt.plot(dp_4_train['avg_actual_loss'], label="width 4")
plt.plot(dp_16_train['avg_actual_loss'], label="width 16")
plt.plot(dp_32_train['avg_actual_loss'], label="width 32")
plt.plot(dp_64_train['avg_actual_loss'], label="width 64")
plt.plot(dp_96_train['avg_actual_loss'], label="width 96")

plt.title("Training Curves for Varying Widths, Weight Decay 0.0001 with DP")
plt.xlabel("epoch")
plt.ylabel('error')
plt.legend()
```
