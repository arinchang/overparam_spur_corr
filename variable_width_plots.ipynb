```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Weight Decay 0.0001 Varying Width Results

```python
wd0001_1 = pd.read_csv("logs/test_1.csv")
wd0001_1
```

```python
wd0001_4 = pd.read_csv("logs/test_2.csv")
wd0001_4
```

```python
wd0001_16 = pd.read_csv("logs/test_3.csv")
wd0001_16
```

```python
wd0001_32 = pd.read_csv("logs/test_4.csv")
wd0001_32
```

```python
wd0001_64 = pd.read_csv("logs/test_5.csv")
wd0001_64
```

```python
wd0001_96 = pd.read_csv("logs/test_6.csv")
wd0001_96
```

```python
w1_worst_group = np.mean(np.array(wd0001_1['avg_loss_group:3'])[40:])
w4_worst_group = np.mean(np.array(wd0001_4['avg_loss_group:3'])[40:])
w16_worst_group = np.mean(np.array(wd0001_16['avg_loss_group:3'])[40:])
w32_worst_group = np.mean(np.array(wd0001_32['avg_loss_group:3'])[40:])
w64_worst_group = np.mean(np.array(wd0001_64['avg_loss_group:3'])[40:])
w96_worst_group = np.mean(np.array(wd0001_96['avg_loss_group:3'])[40:])

```

```python
test_errors_0001 = np.array([0.311255, 0.183662, 0.316266, 0.289399, 0.304003, 0.290738])
worst_group_errors_0001 = np.array([w1_worst_group, w4_worst_group, w16_worst_group, w32_worst_group, w64_worst_group, w96_worst_group])
plt.plot(np.array([1, 4, 16, 32, 64, 96]), test_errors_0001, marker='o', label='average test')
plt.plot(np.array([1, 4, 16, 32, 64, 96]), worst_group_errors_0001, marker='o', label='worst-group test')
plt.title("CelebA with 0.0001 Weight Decay")
plt.xlabel("ResNet Width")
plt.ylabel("Error")
plt.legend()
```
